**TASK-Prediction Using Decision Tree Algorithm**

**Importing the required libraries and the Iris dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
a=datasets.load_iris()
```

**Exploring the dataset**

```python
a
```

```python
a.data
```

**Creating a DataFrame of the dataset**

```python
iris_df=pd.DataFrame(a.data,columns=['sepal_length',
  'sepal_width',
  'petal_length',
  'petal_width'])
```

```python
iris_df
```

```python
iris_df.cov()
```

```python
iris_df.isnull().sum()
```

**Adding the target column in the DataFrame**

```python
iris_df['target']=a.target
```

```python
iris_df
```

```python
iris_df['target'].replace([0,1,2],['setosa', 'versicolor', 'virginica'],inplace=True)
```

```python
iris_df
```

* As we can see that in the dataset , on the basis of various attributes like sepal_length , sepal_width , etc we have to  predict the species of a particular flower instance . 
* So we can say that , in other words , we have to classify the new flower instance as some species .
* This kind of problem is known as Classification problem in ML and here we can make use of the Decision Tree Classifier .

**Decision Tree Model**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score
DT=DecisionTreeClassifier()
```

```python
X=iris_df.iloc[:,:4]
```

```python
Y=iris_df.iloc[:,4]
```

**Splitting the dataset into train and test datasets**

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

**Training the model on the training dataset**

```python
DT=DT.fit(X_train,Y_train)
```

**Visualising the tree**

```python
fig, axes = plt.subplots(figsize = (25,25))
tree.plot_tree(DT,
               feature_names = ['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'], 
               class_names=['setosa', 'versicolor', 'virginica'],
               filled = True);
fig.savefig('imagename.png')
```

**Getting the predicted values for the test dataset**

```python
Y_pred=DT.predict(X_test)
```

**Checking the accuracy of the model**

```python
DT.score(X_test,Y_test)
```

* So , we see that the model performs well on the test dataset giving high accuracy.
